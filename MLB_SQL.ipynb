### Question 1: Explain what the term sampling distribution means? What are sampling distributions used for.
**Ans:** A sample distribution is a probability distribution of randlomly sampled data drawn from a population. For example, we toll 10 dices in each sampling process and repeat this process for 10000 times. The propbability distribution of the obtained sampling data (ie. dice toll outcomes) is the sampling distribution. We use sample distribution to get statistical inference about the population. We can apply central limit theorem (CLT) on the sample distribution to get idea about the population's mean.

### Question 2: Simulation to investigate the sampling distribution

This simulation is done by first creating a list of 10,000 sampling data which has the specified sample size ($n=10$ or $n=30$). Then we can calculate the variance of each sample to get the distribution of the variance.

#### Sample size $n=10$ 

```python
# get 10,0000 samples from the specified normal distribution with sample size of 10
samples_n10 = [npr.normal(loc = 0, scale = 1, size = 10) for _ in range(100000)]

# check the first 5 samples
print(f"The first five samples: {samples_n10[:5]}")

# get the variance
vars_n10 = np.var(samples_n10, axis = 1)

# check variance of the first 5 samples
print(f"Variance of the first five samples: {vars_n10[:5]}")

# plot a histogram of the sample variance
plt.hist(vars_n10, bins = 20)
plt.title("10,0000 Samples with Sample Size = 10")
plt.xlabel("Variance")
plt.ylabel("Counts")
plt.show()

# calculate the probability of samples with variance larger than 2
p_n10 = np.mean(vars_n10 > 2)
md(f"The probability of observing a sample variance larger than 2 is {p_n10 *100}%.")
```

#### Sample size $n=30$ 

```python
# get 10,0000 samples from the specified normal distribution with sample size of 30
samples_n30 = [npr.normal(loc = 0, scale = 1, size = 30) for _ in range(100000)]

# check the first 5 samples
print(f"The first five samples: {samples_n30[:5]}")

# get the variance
vars_n30 = np.var(samples_n30, axis = 1)

# check variance of the first 5 samples
print(f"Variance of the first five samples: {vars_n30[:5]}")

# plot a histogram of the sample variance
plt.hist(vars_n30, bins = 20)
plt.title("10,0000 Samples with Sample Size = 30")
plt.xlabel("Variance")
plt.ylabel("Counts")
plt.show()

# calculate the probability of samples with variance larger than 2
p_n30 = np.mean(vars_n30 > 2)
md(f"The probability of observing a sample variance larger than 2 is {p_n30 *100}%.")
```

### Task 5
Use SQL code and the HallOfFame and Managers tables to return all of the playerIDs for the people that managed for a team that were inducted into the hall of fame. Also, programmatically report the number of such people.

To get the hall of fame managers, we need to get **inner joint** portion of the **HallOfFame** and **Managers** tables, and then filter out those who are inducted with the condition `inducted = "Y"`:

```python
# SQL query
sql_query = '''
	SELECT l.playerID
 	FROM HallOfFame as l
    INNER JOIN Managers as r ON l.playerID = r.playerID
	WHERE l.inducted = "Y";
	'''
 
# execute query and get the results
df = pd.read_sql(sql_query, con)

# get the counts of each selected player
df2 = df.playerID.value_counts().reset_index().set_axis(["playerID", "Counts"], axis = 1)
display(df2)

# report how many unique Hall of Fame managers
md(f"There are {len(df2)} people that managed for a team that were inducted into the hall of fame.")
```

### Task 6
Now use the same two tables and an SQL query to return every season managed by each manager that made it to the hall of fame. You should return the playerID (manager ID), G, W, and L columns from the Managers table.

Second, determine the overall win/loss records (sum of wins and sum of losses) for each of these hall of fame managers. Third, create a new variable that is the win/loss percentage (W/(W+L)). Lastly, sort the resulting data by the win/loss percentage variable (from largest to smallest).

To calculate each player's game statistics, we first group the data by player then do aggregation for the calculations and sorting:

```python
# SQL query
sql_query = '''
	SELECT r.yearID AS Year, l.playerID, r.G, r.W, r.L
    FROM HallOfFame as l
    INNER JOIN Managers as r ON l.playerID = r.playerID
	WHERE l.inducted = "Y";
	'''
 
# execute query and get the results
df = pd.read_sql(sql_query, con)
display(df)

# calculate overall win/loss and win/loss percentage for each hall of fame manager
df2 = df.groupby("playerID").agg(OveralWins = ("W", sum), OveralLosses = ("L", sum))
df2["WL_percentage"] = df2.OveralWins / (df2.OveralWins + df2.OveralLosses)

# sort the data by the win/loss percentage
df3 = df2.sort_values("WL_percentage", ascending = False)
df3
```

### Task 7
Using SQL, construct a table of hall of fame pitchers (any hall of famer that pitched) that gives the playerID and their total (sum) for GS, G, W, L, IPOuts, CG, SHO, and SV columns.

We will use **HallOfFame** and **Pitching** Tables to obtain the Hall of Fame pitchers before we calculate their statistics:

```python
# SQL query
sql_query = '''
	SELECT l.playerID, r.GS, r.G, r.W, r.L, r.IPOuts, r.CG, r.SHO, r.SV
    FROM HallOfFame as l
    INNER JOIN Pitching as r ON l.playerID = r.playerID
	WHERE l.inducted = "Y";
	'''
 
# execute query and get the results
df = pd.read_sql(sql_query, con)

# calculate overall sums of statistics of interets for each hall of fame pitcher
df_pitching = df.groupby("playerID").agg("sum")
display(df_pitching)

# report how many unique Hall of Fame pitchers
md(f"There are {len(df_pitching)} picther that are inducted into the hall of fame.")
```

### Task 8
For all of the hall of fame pitchers, use SQL to create a table of their batting statistics. Namely, the playerID and their total (sum) for AB, R, H, HR, RBI, BB, and SO.

The list of Hall of Fame pitcher is then left joined with **Batting** tables on `playerID` to get their batting data:

```python
# SQL query
sql_query = '''
	SELECT l.playerID, r2.AB, r2.R, r2.H, r2.HR, r2.RBI, r2.BB, r2.SO
    FROM HallOfFame as l
    INNER JOIN Pitching as r ON l.playerID = r.playerID
    LEFT JOIN Batting as r2 ON l.playerID = r2.playerID
	WHERE l.inducted = "Y";
	'''
 
# execute query and get the results
df = pd.read_sql(sql_query, con)

# calculate overall sums of statistics of interets for each hall of fame pitcher
df_batting = df.groupby("playerID").agg("sum")
display(df_batting)

# report how many unique Hall of Fame pitchers we can their batting data
md(f"There are {len(df_batting)} hall of fame picther that we get their batting data.")
```
